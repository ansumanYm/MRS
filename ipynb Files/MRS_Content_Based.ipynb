#### Merging both the datasets

#### Selecting the Relevant Columns

```python
movies_merge = movies_merge[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_merge.head()
```

#### Removing the Null values.

```python
movies_merge.isna().sum()
```

```python
movies_merge = movies_merge.dropna()
```

## Making Tags

```python
movies_merge.iloc[0].genres
```

```python
import ast
```

```python
# Converting string into list 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

### Get Dictionaries.

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies_merge['genres'] = movies_merge['genres'].apply(convert)
```

```python
movies_merge.head()
```

```python
movies_merge['keywords'] = movies_merge['keywords'].apply(convert)
movies_merge.head()
```

### Top 3 Cast.

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies_merge['cast'] = movies_merge['cast'].apply(convert3)
```

```python
movies_merge.head()
```

### Fetching the name of the director.

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_merge['crew'] = movies_merge['crew'].apply(fetch_director)
```

```python
movies_merge.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_merge['cast'] = movies_merge['cast'].apply(collapse)
movies_merge['crew'] = movies_merge['crew'].apply(collapse)
movies_merge['genres'] = movies_merge['genres'].apply(collapse)
movies_merge['keywords'] = movies_merge['keywords'].apply(collapse)
```

```python
movies_merge['overview'] = movies_merge['overview'].apply(lambda x:x.split())
```

Note : We will give a bit more importance to cast and Crew of the movies by adding them twice. This will suit our recommender system.

```python
movies_merge['tags'] = movies_merge['overview'] + movies_merge['genres'] + movies_merge['keywords'] 
+ movies_merge['cast'] + movies_merge['cast'] + movies_merge['crew'] + movies_merge['crew']
```

```python
movies_merge.head()
```

```python
new = movies_merge.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

### Vectorization
Vectorising all the words in the tags.

We need to apply stemmimg to these words.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
new['tags']
```

## Checking the similarity between movies by cosine similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[0:6]
```

```python
new.iloc[1213].title
```

```python
def get_recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distances = similarity[movie_index]
    R_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:11]
    
    for i in R_list:
        print(new.iloc[i[0]].title)
```

## Testing The Result

```python
get_recommend('Avatar')
```

```python
get_recommend('Batman Begins')
```
